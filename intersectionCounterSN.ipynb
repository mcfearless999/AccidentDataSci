```python

def parseSeconds(minutes, seconds):
    seconds = str(seconds.replace(".",""))
    degrees = str(minutes) + "." + seconds
    return degrees

def updateCnt(cnt):
    cntInt = int(cnt[4:])
    code = cnt[:4]
    cntInt = cntInt + 1
    cnt = code+str(cntInt)
    return cnt

class intersection:
    intType_ = 0
    streetNum_ = 0
    intNum_ = 0
    latNum_ = 0
    longNum_ = 0
    roundflag_ = 0
    accidentCount2012 = 0
    accidentCount2013 = 0
    accidentCount2014 = 0
    accidentCount2015 = 0
    accidentCountTotal = 0
    def __init__(self, intType, streetNum,intNum,latNum,longNum,roundflag):
        self.intType_ = intType
        self.streetNum_ = streetNum
        self.intNum_ = intNum
        self.latNum_ = latNum
        self.longNum_ = longNum
        self.roundflag_ = roundflag
 
        
    def toString(self):
        outString = str(self.intType_) + "," + str(self.streetNum_) + "," + str(self.intNum_) + "," + str(self.latNum_) + "," + str(self.longNum_) + "," + str(self.roundflag_) + "," + self.countStr() +"\n"
        return outString
    
    def update(self, year):
        if (int(year) == 2012):
            self.accidentCount2012 += 1
        if (int(year) == 2013):
            self.accidentCount2013 += 1
        if (int(year) == 2014):
            self.accidentCount2014 += 1
        if (int(year) == 2015):
            self.accidentCount2015 += 1
        self.accidentCountTotal += 1
    
    def countStr(self):
        cntStr = str(self.accidentCount2012) + "," +str(self.accidentCount2013) + "," +str(self.accidentCount2014) + "," +str(self.accidentCount2015) + "," +str(self.accidentCountTotal)
        return cntStr
UniqueIntersections = {}
IntersectionCnt= []

for i in range(0,9):
    IntersectionCnt.append(0)

rbCnt = 0
    

#Old Index	Crash Type	Collision Type	Crash Severity	Intersection Type	Crash ID	Crash Year	Latitude Degrees	Latitude Minutes	Latitude Seconds	Longitude Degrees	Longitude Minutes	Longitude Seconds
FileIn = open("accidentlocationsSN.csv", 'r')
FileOut = open("intersectionLocSN.csv", 'w')
Schema = "Intersection Type, Street Number,Nearest street number intersection,Latitude, Longitude,roundabout flag,2012 accidents,2013 accidents,2014 accidents,2015 accidents,Total accidents\n"
for line in FileIn:
    temp = line.split(',')
    if (temp[5].find(" ") == -1):
        year = temp[7][:4]
        #print(year)
        IntersectionCnt[int(temp[5])] += 1
        Lat = temp[8][:2] + " " + parseSeconds(temp[9][:2], temp[10])
        Long = temp[11][:4] + " " + parseSeconds(temp[12][:2], temp[13])
        SN = temp[14].rstrip()
        intersect = temp[15].rstrip()
        rbFlag = temp[16][0]
        coord = Lat+ ","+ Long
        if (int(rbFlag)== 1):
            rbCnt +=1
        key = str(SN) + "," + str(intersect)
        #print(key)
        
        value = UniqueIntersections.get(key)
        #print(tempIntersect.toString())
        if(value is None):
            tempIntersect = intersection( temp[5],SN, intersect, Lat, Long, rbFlag)
            tempIntersect.update(year)
            #defCnt = temp[5] +","+ temp[10][0] + ",1"
            UniqueIntersections[key] = tempIntersect
            #print (defCnt)
        else:
            loc = UniqueIntersections[key]
            loc.update(year)
#print(UniqueIntersections)


FileOut.write(Schema)
for key, value in UniqueIntersections.items():
    outStr = value.toString()
    #print(outStr)
    FileOut.write(outStr)

FileOut.write("\n\n\n\n\n")

i = 0
for num in IntersectionCnt:
    
    cntStr = str(i) + "," + str(num) + '\n'
    i+=1
    #print(cntStr)
    FileOut.write(cntStr)
    
FileOut.write(",,,"+ str(rbCnt))
FileOut.close()
```
