```python
accidentsdf = pd.read_csv('Multnomah.csv',low_memory=False)
usedcol = ['Crash Type','Collision Type','Median Type','Crash Severity','Intersection Type','Crash ID','Crash Month',
           'Crash Year','Nearest Intersecting Street Number','Road Character']
accidentsdf = accidentsdf[usedcol].dropna().reset_index()
# accidentsdf.columns = ['Old Index','Crash Type','Collision Type','Crash Severity','Intersection Type','Crash ID','Crash Year','']
accidentsdf['year'] = accidentsdf['Crash Year'].apply(int)
accidentsdf['month'] = accidentsdf['Crash Month'].apply(int)
accidentsdf['year'] = pd.to_datetime(accidentsdf['year'], format='%Y').dt.year
accidentsdf['month'] = pd.to_datetime(accidentsdf['month'], format='%m').dt.month
del accidentsdf['Crash Year']
del accidentsdf['Crash Month']
timedf = accidentsdf[['year','month','Crash Severity','Median Type']].groupby(['year','month','Crash Severity']).count()
timedf = timedf.reset_index()
timedf['MonthYear']=pd.to_datetime(timedf.assign(day=1)[['year','month','day']]).dt.strftime('%Y-%m')
del timedf['year']
del timedf['month']
timedf.index = timedf['MonthYear']
del timedf['MonthYear']
timedf.reset_index()
timedf.to_csv('accidentsbyyear.csv')
```

```python
timedf.head()
```

```python
severity = [2.0,4.0,5.0]
severityframes = []
for s in severity:
    severityframes.append(timedf[timedf['Crash Severity'] == s].reset_index())
    
for df in severityframes:
    df.MonthYear = pd.to_datetime(df.MonthYear)
    del df['Crash Severity']
    df.columns = ['date','count']
    df.index = df.date
    del df['date']
```
